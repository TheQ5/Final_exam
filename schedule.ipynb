```python
mon = []
tue = []
wed = []
thu = []
fri = []
randomname = []
randomtel = []
randomadd = []
allcontext = []
allnum = []
for j in range(len(nametext)):
    allnum.append(j)

randomnum = random.sample(allnum, 25)
for k in randomnum:
    randomname.append(nametext[k])
    
for k2 in randomnum:
    randomtel.append(teltext[k2])
    
for k3 in randomnum:
    randomadd.append(addresstext[k3])
    
for i in range(25):
    allcontext.append(str(randomnum[i])+"\n"+randomname[i]+"\n"+randomtel[i]+"\n"+randomadd[i])
    
for x in allcontext:
    print(x)
# mon = (randomname[0:5])
# print(randomname)
```
